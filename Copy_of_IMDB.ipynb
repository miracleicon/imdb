```python
data['Genre'].fillna(data['Genre'].mode()[0], inplace=True)
```

```python
best_year = data.groupby("Year")["Rating"].mean().idxmax()
print(f"The year with the best average rating is: {best_year}")
```

```python
data["Duration"] = data["Duration"].astype(str).str.extract(r"(\d+)")

# Convert to integer
data["Duration"] = pd.to_numeric(data["Duration"], errors="coerce")
```

```python
duration_correlation = data["Duration"].corr(data["Rating"])
print(duration_correlation)
```

```python
top_10_movies_overall = data.nlargest(10, "Rating")[["Name", "Year", "Rating"]]
top_10_movies_per_year = data.sort_values(["Year", "Rating"], ascending=[True, False]).groupby("Year").head(10)
print(top_10_movies_per_year)
```

```python
popular_movies_per_year = data.groupby("Year")["Name"].count()
print(popular_movies_per_year)
```

```python
data["Votes"] = data["Votes"].astype(str).str.replace(r"[^\d.]", "", regex=True)
data["Votes"] = pd.to_numeric(data["Votes"], errors="coerce")

# Fill missing votes with 0
data["Votes"].fillna(0, inplace=True)
```

```python
top_voted_movies_overall = data.nlargest(10, "Votes")[["Name", "Year", "Votes"]]
top_voted_movies_per_year = data.sort_values(["Year", "Votes"], ascending=[True, False]).groupby("Year").head(10)

print(top_voted_movies_overall)
print(top_voted_movies_per_year)
```

```python
top_director = data["Director"].value_counts().idxmax()
print(f"The director with the most movies in the dataset is: {top_director}")
```

```python
actors = pd.concat([data["Actor 1"], data["Actor 2"], data["Actor 3"]])
top_actor = actors.value_counts().idxmax()
print(top_actor)
```
